```python
import pandas as pd
import numpy as np
from scipy.stats import poisson
```

```python
df = pd.read_excel("cuad13fil.xlsx")
df = df.sort_values('lambda')
df  
```

```python
df['lambda'] = df['lambda'] # demanda, pedidos por semana
lambd = df['lambda']
x = []  

for i, row in df.iterrows():
    lambd = row['lambda']
    current_x = 0

    while poisson.sf(current_x, lambd) >  0.05:
        current_x += 1
    x.append(current_x)

df['x'] = x
df = df[df['x']>0]
df = df.sort_values('lambda')
df
```
